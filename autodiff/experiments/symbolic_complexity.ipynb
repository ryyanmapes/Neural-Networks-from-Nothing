```python
import numpy as np
from libraries.autodiff_symbolic import *

X = "X"
```

```python
eqn = Add(Square(Add(Square(Variable(X)), Mult(Const(2), Variable(X)))), Const(1))
print(eqn)

eqn_prime = eqn.diff(X)
print(eqn_prime)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

LAYERS = 20

eqn_operations = []
eqn_prime_operations = []
for i in range(LAYERS):
    eqn = Variable(X)
    for j in range(i):
        eqn = Tanh(Add(Mult(Const(j+1), eqn), Const(j+1)))
    eqn_operations.append(eqn.terms())
    eqn_prime_operations.append(eqn.diff(X).terms())

plt.plot(range(len(eqn_operations)), eqn_operations, label='F(x)', linewidth=3)
plt.plot(range(len(eqn_prime_operations)), eqn_prime_operations, label='F\'(x)', linewidth=3)
plt.xlabel('Layers', fontsize=15)
plt.ylabel('Operation Count', fontsize=15)
plt.title('Complexity Scaling of Symbolic Derivation', fontsize=15)
plt.legend(fontsize=15)
# Set x-axis ticks to integers
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
